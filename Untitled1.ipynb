```python
import datetime
from re import *

def PhoneValid(number):
  error = "Invalid phone number."
  if re.match(r"^(\([0-9]{3}\) ?|[0-9]{3}-)[0-9]{3}-[0-9]{3}$", number):
    return number
  else:
    return error

def NameValid(name):
  error = "Invalid name."
  if re.match(r'[A-Za-z]{2,25}( [A-Za-z]{2,25})?', name):
    return name
  else:
    return error

def EmailValid (email):
  error = "Invalid e-mail."
  if re.match(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', email):
    return email
  else:
    return error

def BoolValid(input):
  error = "Invalid type of input."
  if type(input)==bool:
    return input
  else:
    return error

def TimeValid(input):
  error = "Invalid time."
  if re.match(r'^(2[0-3]|[01]?[0-9]):([0-5]?[0-9])$', input):
    return input
  else:
    return error

def cvcValid(input):
  error = "Invalid CVC/CVV"
  if input > 99 and input < 1000:
    return input
  else:
    return error

def CardNumberValid(input):
  error = "Invalid card number."
  if re.match(r'^[1-9][0-9]{15}|[973][0-9]{3}-[0-9]{4}-[0-9]{4}-[0-9]{4}$', input):
    return input
  else:
    return error



class Client:
  __clientnumber=0
  def __init__(self, name, phone):
    Client.__clientnumber += 1
    self.__clientId = Client.__clientnumber
    self.name = name
    self.phone = phone

  @property
  def clientId(self):
    return self.__clientId
  @clientId.setter
  def clientId(self,input):
    self.__clientId = input

  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,input):
    self.__name = NameValid(input)
 
  @property
  def phone(self):
    return self.__phone  
  @name.setter
  def phone(self, input):
    self.__phone = PhoneValid(input)

  def Show_Info(self):
    print("Client's ID:",self.__clientId,"\nClent's name:",self.__name,"\nClient's phone:",self.__phone)

class Manager:
  def __init__(self, name):
    self.name = name
  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,input):
    self.__name=input
  def Show_Info(self):
    print("Manager:",self.__name)

class Organizer:
  def __init__(self, name):
    self.name = name
  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,input):
    self.__name = input
  def Show_Info(self):
    print("Organizer:",self.__name)

class Chef:
  def __init__(self, name):
    self.name = name
  @property
  def name(self):
    return self.name
  @name.setter
  def name(self,input):
    self.__name = NameValid(input)
  def Show_Info(self):
    print("Chef: ", self.__name)

class Supplier:
  __supplierNumber = 0
  def __init__(self, productType, name, phone, email, deliveryTime):
    Supplier.__supplierNumber += 1
    self.__supplierId = Supplier.__supplierNumber
    self.__productType = productType
    self.name = name
    self.phone = phone
    self.email = email
    self.__deliveryTime = deliveryTime

  @property
  def supplierId(self):
    return self.__supplierId
  @supplierId.setter
  def supplierId(self,input):
    self.__supplierId = input

  @property
  def productType(self):
    return self.__productType
  @productType.setter
  def productType(self,input):
    self.__productType = input

  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,input):
    self.__name = NameValid(input)

  @property
  def phone(self):
    return self.__phone
  @phone.setter
  def phone(self,input):
    self.__phone = PhoneValid(input)

  @property
  def email(self):
    return self.__email
  @email.setter
  def email(self,input):
    self.__email = EmailValid(input)

  @property
  def deliveryTime(self):
    return self.__deliveryTime
  @deliveryTime.setter
  def deliveryTime(self,input):
    self.__deliveryTime = input

  def Show_Info(self):
    print("Supplier's ID:",self.__supplierId,"\nProduct dor delivering:",self.__productType,"\nSupplier's name:", self.__name,"\nSupplier's phone:",self.__phone,"\nSupplier's e-mail:",self.__email,"\nDelivery time:",self.__deliveryTime)

class Product:
  __ProductCount = 0
  def __init__(self, name, measurement, quantity, cost):
    Product.__ProductCount += 1
    self.__productId = Product.__ProductCount
    self.name = name
    self.__measurement = measurement
    self.__quantity = quantity
    self.__cost = cost

  @property
  def productId(self):
    return self.__productId
  
  @productId.setter
  def productId(self,input):
    self.__productId = input
  
  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,input):
    self.__name = NameValid(input)

  @property
  def measurement(self):
    return self.__measurement
  @measurement.setter
  def measurement(self,input):
    self.__measurement = input

  @property
  def quantity(self):
    return self.__quantity
  @quantity.setter
  def quantity(self,input):
    self.__quantity = input

  @property
  def cost(self):
    return self.__cost
  @cost.setter
  def cost(self,input):
    self.__cost = input

  def Show_Info(self):
    print("\nProduct's ID:",self.__productId,"\nProduct's name:",self.__name,"\nProduct's measurement:",self.__measurement,"\nProduct's quantity:",self.__quantity,"\nProduct's cost:",self.__cost)

class Ingredient:
  __IngredientCount = 0
  def __init__(self, name, supplier, expireDate):
    Ingredient.__IngredientCount += 1
    self.__ingredientId = Ingredient.__IngredientCount
    self.name = name
    self.__supplier = supplier
    self.__expireDate = expireDate
    
  @property
  def ingredientId(self):
    return self.__ingredientId
  @ingredientId.setter
  def ingredientId(self,input):
    self.__ingredientId = input
    
  @property
  def name(self):
    return self.__name
  @name.setter
  def name(self,input):
    self.__name = NameValid(input)

  @property
  def supplier(self):
    return self.__supplier
  @supplier.setter
  def supplier(self,input):
    self.__supplier = input
    
  @property
  def expireDate(self):
    return self.__expireDate
  @expireDate.setter
  def expireDate(self,input):
    self.__expireDate = input
    
  def Show_Info(self):
    print("\nIngredient's ID:",self.__ingredientId,"\nIngredient's name:", self.name,"\nSupplier's ID:", self.__supplier.supplierId,"\nIngredient's expire date:", self.__expireDate)

    
class Recept:
  __ReceptCount = 0
  def __init__(self,product, ingredient, measurement, quantity):
    Recept.__ReceptCount += 1
    self.__receptId = Recept.__ReceptCount
    self.__product = product
    self.__ingredient = ingredient
    self.__measurement = measurement
    self.__quantity = quantity

  @property
  def ingredient(self):
    return self.__ingredient
  @ingredient.setter
  def ingredient(self,input):
    self.__ingredient = input


  @property
  def receptId(self):
    return self.__receptId
  @receptId.setter
  def receptId(self,input):
    self.__receptId = input


  @property
  def product(self):
    return self.__product
  @product.setter
  def product(self,input):
    self.__product = input


  @property
  def measurement(self):
    return self.__measurement
  @measurement.setter
  def measurement(self,input):
    self.__measurement = input

  @property
  def quantity(self):
    return self.__quantity
  @quantity.setter
  def quantity(self,input):
    self.__quantity = input

  def dicprint(dict):
    for key, value in dict.items(): 
      print(key+" : ", end= value+"\n")

  def Show_Info(self):
    print("\nRecept ID:",self.__receptId,"\nProduct ID:",self.__product.productId)
    Recept.dicprint(self.ingredient)
    print("\nMeasurement of product:",self.__measurement,"\nQuantity of product:",self.__quantity)


class Reservation:
  __ReservationCount = 0
  def __init__(self,client, tableNumber, tableCount, parking, dateArrival, dateTime):
    Reservation.__ReservationCount += 1
    self.__reservationId = Reservation.__ReservationCount
    self.__client = client
    self.__tableNumber = tableNumber
    self.__tableCount = tableCount
    self.parking = parking
    self.__dateArrival = dateArrival
    self.dateTime = dateTime

  @property
  def reservationId(self):
    return self.__reservationId
  @reservationId.setter
  def reservationId(self,input):
    self.__reservationId = input

  @property
  def client(self):
    return self.__client
  @client.setter
  def client(self,input):
    self.__client = input

  @property
  def tableNumber(self):
    return self.__tableNumber
  @tableNumber.setter
  def tableNumber(self,input):
    self.__tableNumber = input

  @property
  def tableCount(self):
    return self.__tableCount
  @tableCount.setter
  def tableCount(self,input):
    self.__tableCount = input

  @property
  def parking(self):
    return self.__parking
  @parking.setter
  def parking(self,input):
    self.__parking = BoolValid(input)

  @property
  def dateArrival(self):
    return self.__dateArrival
  @dateArrival.setter
  def dateArrival(self,input):
    self.__dateArrival = input

  @property
  def dateTime(self):
    return self.__dateTime
  @dateTime.setter
  def dateTime(self,input):
    self.__dateTime = TimeValid(input)

  def Show_Info(self):
    print("\nReservation ID:",self.__reservationId, "\nClient ID:",self.__client.clientId, "\nReserved table number:",self.__tableNumber, "\nQuantity of reserved tables:",self.__tableCount, "\nNeeding of parking:",self.__parking, "Date of reserve:",self.__dateArrival, "Time of reserve:",self.__dateTime)

class OrderDetail:
  __OrderDetailCount = 0
  def __init__(self, reservation, product):
    OrderDetail.__OrderDetailCount += 1
    self.__orderDetailId = OrderDetail.__OrderDetailCount
    self.__reservation = reservation
    self.__product = product

  @property
  def orderDetailId(self):
    return self.__orderDetailId
  @orderDetailId.setter
  def orderDetailId(self,input):
    self.__orderDetailId = input

  @property
  def reservation(self):
    return self.__reservation
  @reservation.setter
  def reservation(self,input):
    self.__reservation = input

  @property
  def product(self):
    return self.__product
  @product.setter
  def product(self,input):
    self.__product = input
  def dicprint1(dict):
    for key, value in dict.items(): 
      print( key ,value, end="\n")
  def Show_Info(self):
    print("\nOrder Detail ID:",self.__orderDetailId, "\nReservation ID:",self.__reservation.reservationId)
    OrderDetail.dicprint1(self.__product)

class ClientForm:
  __ClienFormCount = 0
  def __init__(self, reservation, orderDetail, payment):
    ClientForm.__ClienFormCount += 1
    self.__clientFormId = ClientForm.__ClienFormCount
    self.__reservation = reservation
    self.__orderDetail = orderDetail
    self.__payment = payment


  @property
  def clientFormId(self):
    return self.__clientFormId
  @clientFormId.setter
  def clientFormId(self,input):
    self.__clientFormId = input


  @property
  def reservation(self):
    return self.__reservation
  @reservation.setter
  def reservation(self,input):
    self.__reservation = input

  @property
  def orderDetail(self):
    return self.__orderDetail
  @orderDetail.setter
  def orderDetail(self,input):
    self.__orderDetail = input

  
  @property
  def payment(self):
    return self.__payment
  @payment.setter
  def payment(self,input):
    self.__payment = input
  def Show_Info(self):
    print("\nClient Form ID:",self.__clientFormId, "\nOrder Detail ID:",self.__orderDetail.orderDetailId, "\nReservation ID:",self.__reservation.reservationId, "\nPayment ID:",self.__payment.paymentId)

     
class Payment:
  __PaymentCount = 0
  def __init__(self, firstName, lastName, cardNumber, cvc, allTheCost):
    Payment.__PaymentCount += 1
    self.__paymentId = Payment.__PaymentCount
    self.__firstName = firstName
    self.__lastName = lastName
    self.cardNumber = cardNumber
    self.cvc = cvc
    self.__allTheCost = allTheCost
  
  @property
  def paymentId(self):
    return self.__paymentId
  @paymentId.setter
  def paymentId(self,input):
    self.__paymentId = input

  @property
  def firstName(self):
    return self.__firstName
  @firstName.setter
  def firstName(self,input):
    self.__firstName = input

  @property
  def lastName(self):
    return self.__lastName
  @lastName.setter
  def lastName(self,input):
    self.__lastName = input

  @property
  def cardNumber(self):
    return self.__cardNumber
  @cardNumber.setter
  def cardNumber(self,input):
    self.__cardNumber = CardNumberValid(input)
 
  @property
  def cvc(self):
    return self.__cvc
  @cvc.setter
  def cvc(self,input):
    self.__cvc = cvcValid(input)

  @property
  def allTheCost(self):
    return self.__allTheCost
  @allTheCost.setter
  def allTheCost(self,input):
    self.__allTheCost = input

  def Show_Info(self):
    print("\nPayment ID:",self.__paymentId, "\nPayer's First Name:",self.__firstName,"\nPayer's Last Name:", self.__lastName,"\nCard Numbers:", self.__cardNumber,"\nCVC2/CVV2:", self.__cvc, "\nThe final cost:",self.__allTheCost)


client_john = Client("John" , "077-844-444")
client_jean = Client("Jean", "091-252-425")
Client_list = [client_john, client_jean]

manager = Manager("Natalie")
Manager_list = [manager]

org = Organizer("Malena")
Organizer_list = [org]

chef = Chef("Grisha")
Chef_list = [chef]

sup_meat = Supplier("meat", "Misaq", "077-289-673", "meatsupplier@gmail.com", '20th of the month')
sup_drink = Supplier("drinks", "Aleksan", "093-202-020", "aleksan.vodka@yahoo.com", "21th of the month")
sup_bread = Supplier("bread and flour", "BreadFactory", "055-202-127", "factorybread@yahoo.com", '22th of the month')
sup_dairy = Supplier("dairy products", "Marianna", "010-242-424", 'marianna.dairy@gmail.com', '23th of the month')
sup_market = Supplier("mixed products", "Market", '055-200-201', 'market@yandex.ru', '24th of the month')
Supplier_list = [sup_meat, sup_drink, sup_bread, sup_dairy, sup_market]

pizza=Product('margarita','peace',1,5)
juice=Product('Cola','litre',1,7)
salad=Product('ceaser','portion',1,10)
Product_list = [pizza, juice, salad]

product_john = {pizza.name:4, juice.name:2, salad.name:1}
product_jean = {salad.name:4, juice.name:10, pizza.name:2}

reservation_john = Reservation(client_john,4,1,True,'15-11-2021','19:00')
reservation_jean = Reservation(client_jean, '7-9', 2, False, '12-11-2021', '18:00')
Reservation_list = [reservation_john, reservation_jean]

orderdetail_john = OrderDetail(reservation_john,product_john)
orderdetail_jean = OrderDetail(reservation_jean, product_jean)
Order_detail_list = [orderdetail_john, orderdetail_jean]

ingredient_meat = Ingredient('meat',sup_meat,'1 year')
ingredient_sausage = Ingredient('sausage',sup_meat,'15 days')
ingredient_bread = Ingredient('bread',sup_bread,'1 day')
ingredient_dough = Ingredient('dough',sup_bread,'1 year')
ingredient_cheese = Ingredient('cheese',sup_dairy,'3 months')
ingredient_olivie = Ingredient('olivie',sup_market,'1 year')
ingredient_sauce = Ingredient('sauce',sup_market,'2 months')
ingredient_greens = Ingredient('greens',sup_market,'3 days')
ingredient_cola = Ingredient('cola',sup_drink,'2 years')
ingredient_water = Ingredient('water',sup_drink,'1 year')
Ingredient_list = [ingredient_meat, ingredient_sausage, ingredient_bread, ingredient_dough, ingredient_cheese, ingredient_olivie, ingredient_sauce, ingredient_greens, ingredient_cola, ingredient_water]


pizza_ingr = {ingredient_dough.name: "300 gram", ingredient_sausage.name: "100 gram", ingredient_cheese.name: "250 gram", ingredient_sauce.name: "100 ml",ingredient_olivie.name: "50 gram"}
juice_ingr = {ingredient_cola.name: "500 ml"}
ceaser_ingr = {ingredient_bread.name: "100 gram", ingredient_meat.name: "200 gram", ingredient_cheese.name: "60 gram",ingredient_greens.name: "100 gram", ingredient_sauce.name: "150 ml"}

recept_pizza = Recept( pizza,pizza_ingr, "peace", 6)
recept_cola = Recept( juice, juice_ingr, "ml", 500)
recept_ceaser = Recept( salad, ceaser_ingr, "portion", 1)
Recept_list = [recept_pizza, recept_cola, recept_ceaser]

payment_john = Payment('JOHN','KARTAPETYAN','1234567890204875',456,12000)
payment_jean = Payment('JEAN', 'ASRIBABAYAN','9876543210123477', 321, 15000)
Payment_list = [payment_john, payment_jean]

client_form_john = ClientForm(reservation_john,orderdetail_john,payment_john)
client_form_jean = ClientForm(reservation_jean,orderdetail_jean,payment_jean)
ClientForm_list = [client_form_john, client_form_jean]


```
